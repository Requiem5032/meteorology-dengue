```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
calibration_df = pd.read_csv('data/calibration/bello/data.csv')
projection_df = pd.read_csv('data/projection/bello/data.csv')
```

```python
calibration_cases = [calibration_df['cases'][0]]
for i in range(1, len(calibration_df['cases'])):
    calibration_cases.append(calibration_df['cases'][i] - calibration_df['cases'][i-1])

projection_cases = [projection_df['cases'][0]]
for i in range(1, len(projection_df['cases'])):
    projection_cases.append(projection_df['cases'][i] - projection_df['cases'][i-1])
```

```python
plt.plot(calibration_cases, label='Calibration')
plt.plot(projection_cases, label='Projection')
plt.legend()
plt.title('Outbreak comparison: Calibration vs Projection')
```
